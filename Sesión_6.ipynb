# **Else** De otro modo

 esta instrucción da la acción a realizar en caso que la condición en el if o en elsif (Si anidado) no se cumpliera.

 Va acompañado de dos puntos **else:**

```python
# Ejemplo 1: Número par o impar
numero = 7

if numero % 2 == 0:              #Si el residuo al dividir por 2 es distinto es cero
    print(f"{numero} es par")    #entonces es un número par
else:                            #De lo contrario
    print(f"{numero} es impar")  #Es un número impar
```

```python
# Ejemplo 2: Aprobado o reprobado
calificación = 85

if calificación >= 70:     #Si la variable "Calificación es mayor o igual que 70"
    print("¡Aprobado!")    #Entonces Aprobado
else:                      #De lo contrario
    print("¡Reprobado!")   #Reprobado
```

# **Operadores lógicos and y or**

Son comparativos de al menos dos condiciones para poder ejecutar un bloque de código

and: Evalua las condiciones y en todas deben arrojar True

or: Evalua que al menos una de las condiciones arroje un true

not: Evalua que se cumpla una sentecia negativa para poder continuar al siguiente paso, es decir busca un false

```python
# AND (y) - Ambas condiciones deben ser True
edad = 25
tiene_licencia = True

if edad >= 18 and tiene_licencia:
    print("Puedes conducir")
else:
    print("Toco en metro")

# OR (o) - Al menos una condición debe ser True
es_fin_de_semana = False
es_feriado = True

if es_fin_de_semana or es_feriado:
    print("No hay clases")
else:
    print("Hay clases")

# NOT (no) - Invierte el resultado
es_admin = False

if not es_admin:
    print("Acceso limitado")
else:
    print("Acceso total")
```

# **El ciclo While (Mientras)**

El objetivo de este ciclo es repetir un bloque de codigo mientras una condición sea True, tiene un riesgo y es que si la condición nunca se vuelve false, el programa se ejecutra hasta que se cuelgue

```python
#Cuenta Regresiva la prueba se volvera false en el momento que sea un valor negativo

contador = 5

while contador > 0:
    print(contador)
    contador = contador - 1 # ¡La línea de escape! Sin esto, es un ciclo infinito.

print("¡Despegue!")
```

#** El ciclo for (Para cada)**

Este ciclo itera sobre una secuencia, de elementos almacenados en una variable, y va a hacer  una acción sobre cada una de ellas o sobre el segmento que se delimite de ellas.

```python
frutas = ["Manzana", "Pera", "Naranja", "Banana"]

for una_fruta in frutas:
    print(f"Me voy a comer una {una_fruta}")
```

La función **Range** es un buen complemento del ciclo for, pues permite crear listas de elementos que pueden interactuar con las variables del ciclo

```python
#Tabla del 7
numero = 7
for i in range(1, 11): # Del 1 al 10
    print(f"{numero} x {i} = {numero * i}")
```

#**Contoladores de ciclos (break y continue)**

**break** termina el ciclo inmediatamente

**continue** contninua a la siguiente iteración del ciclo, ignorando el resto del código de esa vuelta

```python
# Buscar un número en una lista
numeros = [5, 8, 3, 12, 7, 9]
buscar = 12

for numero in numeros:
    print(f"Revisando el número: {numero}")
    if numero == buscar:
        print("¡Número encontrado!")
```

```python
# Buscar un número en una lista y salir cuando lo encuentre
numeros = [5, 8, 3, 12, 7, 9]
buscar = 12

for numero in numeros:
    print(f"Revisando el número: {numero}")
    if numero == buscar:
        print("¡Número encontrado!")
        break  # Sale del ciclo inmediatamente

print("Búsqueda terminada")
```

```python
for num in range(1, 11):
    if num % 2 != 0: # Si el número es impar...
          continue   # ...sáltatelo y ve al siguiente número
    print(num) # Esta línea solo se ejecuta para pares
```

#**Crear una Calculadora basica en la consola**



Intentar (try) pedir al usuario el primer número input.

Corregir (except) un ValueError si no escribe un número válido.

Pedir la operación (ej. "+", "-", "*", "/").

Intentar (try) pedir el segundo número input.

Capturar (except) un ValueError si no es válido.

Usar un bloque if/elif/else para comprobar qué operación eligió.

Intentar (try) realizar el cálculo.

Capturar (except) un ZeroDivisionError si el usuario intentó dividir por cero.

Mostrar el resultado.



```python
print("¡Bienvenido a la calculadora básica!")
while True:
    try:
        num1 = float(input("Ingresa el primer número: "))
    except ValueError:
        print("Por favor, ingresa un número válido.\n")
        continue

    operacion = input("Ingresa la operación (+, -, *, /): ")

    if operacion not in ["+", "-", "*", "/"]:
        print("Operación no válida. Intenta de nuevo.\n")
        continue

    try:
        num2 = float(input("Ingresa el segundo número: "))
    except ValueError:
        print("Por favor, ingresa un número válido.\n")
        continue

    try:
        if operacion == "+":
            resultado = num1 + num2
        elif operacion == "-":
            resultado = num1 - num2
        elif operacion == "*":
            resultado = num1 * num2
        elif operacion == "/":
            resultado = num1 / num2  # Aquí puede ocurrir ZeroDivisionError

        print(f"✅ Resultado: {num1} {operacion} {num2} = {resultado}\n")

    except ZeroDivisionError:
        print("Error: No se puede dividir por cero.\n")
        continue

    respuesta = input("¿Deseas realizar otra operación? (sí/no): ").lower()
    if respuesta != 'sí' and respuesta != 'si':
        print("¡Gracias por usar la calculadora!")
        break
```
